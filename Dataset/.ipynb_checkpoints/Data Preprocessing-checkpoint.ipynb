```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from scipy import stats
from sklearn.model_selection import train_test_split
```

```python
file_path = 'South_East_Asia_Social_Media_MentalHealth.csv'
data = pd.read_csv(file_path)

data.head()
```

```python
# 1. Checking for missing values
print("Missing Values in each column:")
data.isnull().sum()
```

```python
# Handling missing values by imputing median for numerical columns
for col in data.select_dtypes(include=['float64', 'int64']).columns:
    data[col] = data[col].fillna(data[col].median())
```

```python
# Handling missing values by imputing mode for categorical columns
for col in data.select_dtypes(include=['object']).columns:
    data[col] = data[col].fillna(data[col].mode()[0])
```

```python
print("Missing Values After Handling:")
data.isnull().sum()
```

```python
# 2. Detecting and Handling Outliers
# Use boxplots to visualize outliers in numerical columns
def detect_outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = series[(series < lower_bound) | (series > upper_bound)]
    return outliers
```

```python
# Numerical columns
numerical_columns = ['Daily SM Usage (hrs)', 'Likes Received (per post)', 'Comments Received (per post)', 
                     'Shares Received (per post)', 'Peer Comparison Frequency (1-10)', 
                     'Social Anxiety Level (1-10)', 'Body Image Impact (1-10)', 
                     'Sleep Quality Impact (1-10)', 'Self Confidence Impact (1-10)', 
                     'Cyberbullying Experience (1-10)', 'Anxiety Levels (1-10)']
```

```python
# Create subplots
n_cols = 3
n_rows = (len(numerical_columns) // n_cols) + (len(numerical_columns) % n_cols > 0)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(10, 2 * n_rows))
axes = axes.flatten()

for i, col in enumerate(numerical_columns):
    if col in data.columns:
        sns.boxplot(x=data[col], ax=axes[i])
        axes[i].set_title(f'Boxplot for {col}')
        
        outliers = detect_outliers_iqr(data[col].dropna())
        print(f"{col} has {len(outliers)} outliers")

plt.tight_layout()
plt.show()
```

```python
# Handle outliers using IQR if there are any
for col in numerical_columns:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[col] = np.where(data[col] < lower_bound, lower_bound, data[col])
    data[col] = np.where(data[col] > upper_bound, upper_bound, data[col])
```

```python
# Save the preprocessed data
data.to_csv('Preprocessed_Dataset.csv', index=False)
```

```python
# 3. Data Transformation
#Categorize the 'Age Group' column into 'Youth', 'Adult', and 'Senior' categories
def categorize_age_group(age_group):
    if age_group in ['18-25', '26-35']:
        return 'Youth'
    elif age_group in ['36-45', '46-55']:
        return 'Adult'
    else:
        return 'Senior'

```

```python
data['Age Category'] = data['Age Group'].apply(categorize_age_group)
print("\nDataset with 'Age Category' column:")
data[['Age Group', 'Age Category']].head()
```

```python
# 3. Data Transformation
#Combine 'Likes Received (per post)', 'Comments Received (per post)', and 'Shares Received (per post)' into 'Total Social Interaction'
data['Total Social Interaction'] = data['Likes Received (per post)'] + data['Comments Received (per post)'] + data['Shares Received (per post)']
data[['Likes Received (per post)', 'Comments Received (per post)', 'Shares Received (per post)', 'Total Social Interaction']].head()
```

```python
# 3. Data Transformation
# Create a new column 'Usage Intensity' based on 'Daily SM Usage (hrs)'
def usage_intensity(hours):
    if hours < 2:
        return 'Low'
    elif hours < 5:
        return 'Medium'
    else:
        return 'High'
```

```python
data['Usage Intensity'] = data['Daily SM Usage (hrs)'].apply(usage_intensity)

print("\nDataset with 'Usage Intensity' column:")
data[['Daily SM Usage (hrs)', 'Usage Intensity']].head()
```

```python
# 3. Data Transformation
# Create a new column 'Usage Anxiety Interaction' by multiplying 'Daily SM Usage (hrs)' and 'Social Anxiety Level (1-10)'
data['Usage Anxiety Interaction'] = data['Daily SM Usage (hrs)'] * data['Social Anxiety Level (1-10)']
```

```python
# Discretize the 'Usage Anxiety Interaction' column into categories (low, medium, high)
bins = [0, 10, 30, 60, 100, float('inf')] 
labels = ['low', 'medium', 'high', 'very_high', 'extreme']  
data['Usage Anxiety Interaction Category'] = pd.cut(data['Usage Anxiety Interaction'], bins=bins, labels=labels)
```

```python
print("\nDataset with 'Usage Anxiety Interaction' and 'Usage Anxiety Interaction Category' columns:")
data[['Daily SM Usage (hrs)', 'Social Anxiety Level (1-10)', 'Usage Anxiety Interaction', 'Usage Anxiety Interaction Category']].head()
```

```python
#3.Data Transformation-Drop Column
columns_to_drop = ['State', 'Age Group', 'Most Used SM Platform', 'Likes Received (per post)', 'Comments Received (per post)', 
                     'Shares Received (per post)']
```

```python
data = data.drop(columns=columns_to_drop)
```

```python
data.head()
```

```python
# 3. Data Transformation
# Data Encoding for Categorical Variables
categorical_columns = ['Country', 'Gender', 'Urban/Rural', 'Frequency of SM Use', 'Socioeconomic Status', 'Education Level',
                       'Age Category','Usage Intensity','Usage Anxiety Interaction Category']
```

```python
encoder = LabelEncoder()

for col in categorical_columns:
    data[col] = encoder.fit_transform(data[col])

    print(f"\nEncoding for '{col}':")
    encoding_map = pd.DataFrame({
        'Original': encoder.classes_,  
        'Encoded': range(len(encoder.classes_)) 
    })
    print(encoding_map)

```

```python
print("\nDataset after Encoding:")
data.head()
```

```python
#Data Splitting
# Splitting the original dataset into training (80%) and validation (20%)
train_data, validation_data = train_test_split(data, test_size=0.2, random_state=42)
```

```python
# Further splitting the training data into 80% training and 20% testing
train_data, test_data = train_test_split(train_data, test_size=0.2, random_state=42)
```

```python
print("Training Dataset Size:", train_data.shape)
print("Validation Dataset Size:", validation_data.shape)
print("Testing Dataset Size:", test_data.shape)
```

```python
train_data.to_csv("training_data.csv", index=False)
validation_data.to_csv("validation_data.csv", index=False)
test_data.to_csv("testing_data.csv", index=False)

print("Datasets have been saved to 'training_data.csv', 'validation_data.csv', and 'testing_data.csv'.")
```
