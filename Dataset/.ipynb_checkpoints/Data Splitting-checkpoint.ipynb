```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
training_data_path = 'training_data.csv'
#validation_data_path = 'validation_data.csv'
#testing_data_path = 'testing_data.csv'

# Load the data for each set
training_data = pd.read_csv(training_data_path)
validation_data = pd.read_csv(validation_data_path)
testing_data = pd.read_csv(testing_data_path)

# Now you can check or process the data
print("Training Data Head:")
training_data.head()

print("Validation Data Head:")
validation_data.head()

print("Testing Data Head:")
testing_data.head()

```

```python
# Check distributions of a numerical feature like 'Daily SM Usage (hrs)'
plt.figure(figsize=(12, 6))

# Plot distributions across the training, validation, and test sets
sns.histplot(training_data['Daily SM Usage (hrs)'], color='blue', label='Training Set', kde=True, stat='density')
sns.histplot(validation_data['Daily SM Usage (hrs)'], color='orange', label='Validation Set', kde=True, stat='density')
sns.histplot(testing_data['Daily SM Usage (hrs)'], color='purple', label='Test Set', kde=True, stat='density')

plt.legend()
plt.title('Comparison of Daily SM Usage (hrs) Distribution Across Sets')
plt.xlabel('Daily SM Usage (hrs)')
plt.ylabel('Density')
plt.show()
```
