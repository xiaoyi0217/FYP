```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
training_data_path = 'train_data.csv'
validation_data_path = 'val_data.csv'
testing_data_path = 'test_data.csv'

training_data = pd.read_csv(training_data_path)
validation_data = pd.read_csv(validation_data_path)
testing_data = pd.read_csv(testing_data_path)

print("Training Data Head:")
training_data.head().T

print("Validation Data Head:")
validation_data.head().T

print("Testing Data Head:")
testing_data.head().T

```

```python
plt.figure(figsize=(12, 6))

# Plot distributions across the training, validation, and test sets
sns.histplot(training_data['Cyberbullying Experience (1-10)'], color='blue', label='Training Set', kde=True, stat='density')
sns.histplot(validation_data['Cyberbullying Experience (1-10)'], color='orange', label='Validation Set', kde=True, stat='density')
sns.histplot(testing_data['Cyberbullying Experience (1-10)'], color='purple', label='Test Set', kde=True, stat='density')

plt.legend()
plt.title('Comparison of Cyberbullying Experience (1-10) Distribution Across Sets')
plt.xlabel('Cyberbullying Experience (1-10)')
plt.ylabel('Density')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define a list of datasets and their corresponding titles
datasets = [("Training Data", training_data), 
            ("Validation Data", validation_data), 
            ("Testing Data", testing_data)]

# Create a figure to hold the plots
plt.figure(figsize=(18, 5))

# Loop over each dataset and plot the distribution
for i, (title, data) in enumerate(datasets, 1):
    plt.subplot(1, 3, i)  # Create a subplot in a 1x3 grid
    sns.countplot(x='Social_Anxiety_Category', data=data)
    plt.title(f'Class Distribution in {title}')

# Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
```
